```python
import pandas as pd
from sklearn.metrics import accuracy_score, auc, classification_report, confusion_matrix, f1_score, precision_score, recall_score, roc_curve
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

df = pd.read_csv('sonar_dataset.csv')

total_columns = len(df.columns)
target_index = total_columns - 1
target = df.iloc[:, target_index]
features = df.drop(columns=df.columns[target_index])

mapping = {'R': 0, 'M': 1}
target = target.map(mapping)

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)


param_grid = {
    'pca__n_components': [2, 5, 7, 10, 12],
    'classifier__max_depth': [3, 5, 10, 15],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4],
    'classifier__criterion': ['gini', 'entropy'],
    'classifier__max_features': ['sqrt', 'log2'],
    'classifier__min_impurity_decrease': [0.0, 0.01, 0.05, 0.1]
}

pipeline = Pipeline([
    ('scaler', StandardScaler()),           
    ('pca', PCA()),          
    ('classifier', DecisionTreeClassifier())
])

print("Starting GridSearch...")

best = 0
best_cv = 0
best_report = None
best_sensitivity = None
best_specificity = None
best_estimator = None

for cv in range(2 ,20):
    grid_search = GridSearchCV(
        pipeline, 
        param_grid, 
        cv=cv,
        scoring='roc_auc',
        n_jobs=-1,
        verbose=1
    )
    print("\n Current cv: ", cv)
    grid_search.fit(X_train, y_train)

    y_pred = grid_search.predict(X_test)

    test_score = grid_search.best_estimator_.score(X_test, y_test)

    report = classification_report(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    TN, FP, FN, TP = cm.ravel()

    print(f"True Positives (TP): {TP}")
    print(f"True Negatives (TN): {TN}")
    print(f"False Positives (FP): {FP}")
    print(f"False Negatives (FN): {FN}")

    sensitivity = TP / (TP + FN)
    print(f"Sensitivity (Recall): {sensitivity:.4f}")

    specificity = TN / (TN + FP)
    print(f"Specificity: {specificity:.4f}")

    y_proba_1 = grid_search.best_estimator_.predict_proba(X_test)[:, 1]
    y_proba_0 = grid_search.best_estimator_.predict_proba(X_test)[:, 0]

    print("Mine Probability: ",y_proba_1)
    print("Rock Probability: ",y_proba_0)

    fpr, tpr, thresholds = roc_curve(y_test, y_proba_1)
    roc_auc = auc(fpr, tpr)

    score = accuracy_score(y_test, y_pred)
    conf_mtx = confusion_matrix(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)

    print(f"Test Set Score: {test_score:.4f}")    
    print("F1- Score: ", f1)
    print("Recall: ",recall)
    print("Precision: ", precision)

    print("\nClassification Report:")
    print(report)

    print(f"AUC Score For Mine: {roc_auc:.4f}")


    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random classifier')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (1 - Specificity)')
    plt.ylabel('True Positive Rate (Sensitivity)')
    plt.title('Receiver Operating Characteristic (ROC) Curve For Target = 1 (Mine)')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()

    fpr, tpr, thresholds = roc_curve(y_test, y_proba_0)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random classifier')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (1 - Specificity)')
    plt.ylabel('True Positive Rate (Sensitivity)')
    plt.title('Receiver Operating Characteristic (ROC) Curve For Target = 0 (Rock)')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()

    print(f"AUC Score: {roc_auc:.4f}")

    if test_score > best:
        best = test_score
        best_cv = cv
        best_report, best_estimator = report, grid_search.best_estimator_
        best_sensitivity, best_specificity = sensitivity, specificity

print("Best CV_SCORE: ", best_cv , "_", best)
print("Best Estimators: ", best_estimator)
print("Best Accuracy: ", best)
print("Best report: " , best_report)
print("Best sensitivity: " , best_sensitivity)
print("Best specificity: " , best_specificity)

```
